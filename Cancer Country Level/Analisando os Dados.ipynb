```python
import pandas as pd
dados_new = pd.read_csv('dados_death.csv')
```

```python
dados_new
```

Ainda há valores nulos nas colunas 'Tendência Recente' e 'Tendência recente taxas de mortalidade (5 anos)', precisamos excluir essas ocorrências.

```python
selecao = (dados_new['Tendência Recente'] == '**') | (dados_new['Tendência recente taxas de mortalidade (5 anos)'] == '**')
```

```python
dados_new[selecao]
```

```python
dados = dados_new[~selecao]
```

A coluna 'Tendência recente taxas de mortalidade (5 anos)' está sendo linda com o object, precisamos converte-la em int.

Por enquanto, veremos quais as ocorrências que não atingiram o objetivo, mas estavam com tendência recente caindo.

```python
selecao = (dados['Objetivo Taxa Mortalidade (45,5)'] == 'No') & (dados['Tendência Recente'] == 'falling')
```

```python
dados[selecao]
```

```python
dados['Tendência recente taxas de mortalidade (5 anos)'] = dados['Tendência recente taxas de mortalidade (5 anos)'].astype(float)
```

```python
dados['Tendência recente taxas de mortalidade (5 anos)'].dtypes
```

Agora sim, podemos fazer todo o filtro:

```python
selecao = (dados['Objetivo Taxa Mortalidade (45,5)'] == 'No') & (dados['Tendência Recente'] == 'falling') & (dados['Tendência recente taxas de mortalidade (5 anos)'] < 0)
```

```python
dados[selecao]
```

### Quais condados atingiram o objetivo na Taxa de Mortalidade, mas estavam com a Tendencia Recente subindo e a taxa de mortalidade positiva?

```python
dados.info()
```

```python
selecao = (dados['Objetivo Taxa Mortalidade (45,5)'] == 'Yes') & (dados['Tendência Recente'] == 'rising') & (dados['Tendência recente taxas de mortalidade (5 anos)'] > 0)
```

```python
dados[selecao]
```

```python
dados['Média de Mortes/Ano'] = dados['Média de Mortes/Ano'].replace(',', '.', regex=True)
```

```python
dados['Média de Mortes/Ano'] = dados['Média de Mortes/Ano'].astype(float)
```

```python
dados.info()
```

### Quais os condados que mais morrem pessoas por cancer?

```python
selecao = dados[['Condado', 'Média de Mortes/Ano']]
```

```python
selecao.sort_values(by = 'Média de Mortes/Ano', ascending = False).iloc[0:10]
```
